1.3 Generate a Fibonacci sequence of length 5 to project the next 5 years of growth.

```python
# Alternative simpler approach
def simple_fibonacci_projection(actual_data, years=5):
    """Simple Fibonacci-based projection"""
    # Generate Fibonacci sequence
    fib = [1, 1, 2, 3, 5]  # Length 5 Fibonacci sequence
    
    # Calculate recent average growth rate
    growth_rates = []
    for i in range(1, len(actual_data)):
        rate = (actual_data[i] - actual_data[i-1]) / actual_data[i-1]
        growth_rates.append(rate)
    
    avg_growth_rate = sum(growth_rates[-3:]) / 3  # Average of last 3 growth rates
    
    # Project using Fibonacci-scaled growth
    current_pop = actual_data[-1]
    projected = []
    
    for i, fib_num in enumerate(fib):
        # Scale growth by Fibonacci number
        growth_factor = avg_growth_rate * (fib_num / max(fib))
        next_pop = current_pop * (1 + growth_factor)
        projected.append(round(next_pop))
        current_pop = next_pop
    
    return fib, projected

print("\nFibonacci sequence of length 5 to project the next years of growth:")
print("=" * 70)

kampala_fib_simple, kampala_proj_simple = simple_fibonacci_projection(kampala)
wakiso_fib_simple, wakiso_proj_simple = simple_fibonacci_projection(wakiso)
gulu_fib_simple, gulu_proj_simple = simple_fibonacci_projection(gulu)

print(f"Kampala projected: {kampala_proj_simple}")
print(f"Wakiso projected: {wakiso_proj_simple}")
print(f"Gulu projected: {gulu_proj_simple}")
```

2.2 Use scipy.linalg.solve() to calculate energy from solar panels and Batteries.

```python
import numpy as np
from scipy import linalg

def solve_energy_system(D1, D2):
    """
    Solves the system of equations to find energy from solar panels (X) and batteries (Y).
    Equations:
    3X + 2Y = D1
    4X +  Y = D2
    """
    # Coefficient matrix
    a = np.array([[3, 2],
                  [4, 1]])
    
    # Right-hand side vector (daily demands)
    b = np.array([D1, D2])
    
    try:
        # Solve the system of linear equations
        x = linalg.solve(a, b)
        return x
    except ValueError as e:
        print(f"Error with input values: {e}")
        return None
    except np.linalg.LinAlgError as e:
        print(f"Linear algebra error: {e}")
        return None

# Get user input for daily demands
print("Enter daily demand values:")
D1 = float(input("D1 (for equation 3X + 2Y) = "))
D2 = float(input("D2 (for equation 4X +  Y) = "))

# Calculate the solution
solution = solve_energy_system(D1, D2)

if solution is not None:
    X, Y = solution
    print(f"\n=== Solution ===")
    print(f"Energy from solar panels (X): {X:.2f} kWh")
    print(f"Energy from batteries (Y): {Y:.2f} kWh")
    
    # Verify the solution
    print(f"\n=== Verification ===")
    print(f"Equation 1: 3*{X:.2f} + 2*{Y:.2f} = {3*X + 2*Y:.2f} (Target: {D1})")
    print(f"Equation 2: 4*{X:.2f} + {Y:.2f} = {4*X + Y:.2f} (Target: {D2})")
```

```python
# Initialize the first two Fibonacci numbers
prev, curr = 0, 1
fib_sequence = [prev, curr]  # Starting the sequence with F0 and F1

# Generate the next 13 Fibonacci numbers to have a total of 15
for _ in range(13):
    prev, curr = curr, prev + curr
    fib_sequence.append(curr)

print("The first 15 Fibonacci numbers are:")
print(fib_sequence)
```

```python
import numpy as np
import math

def cosine_similarity(vector_a, vector_b):
    """
    Compute cosine similarity between two vectors using math.cos()
    Cosine similarity = cos(θ) = (A·B) / (||A|| * ||B||)
    """
    # Calculate dot product A·B
    dot_product = sum(a * b for a, b in zip(vector_a, vector_b))
    
    # Calculate magnitudes (||A|| and ||B||)
    magnitude_a = math.sqrt(sum(a * a for a in vector_a))
    magnitude_b = math.sqrt(sum(b * b for b in vector_b))
    
    # Avoid division by zero
    if magnitude_a == 0 or magnitude_b == 0:
        return 0
    
    # Calculate cosine similarity
    cosine_sim = dot_product / (magnitude_a * magnitude_b)
    
    return cosine_sim

# Original Kampala data
kampala_rainfall = np.array([120, 140, 180, 200, 220, 180, 90, 70, 60, 100, 110, 130])

# Your specific Gulu data
gulu_rainfall = np.array([79, 80, 147, 154, 184, 173, 131, 129, 84, 150, 115, 97])

# Display the data
print("Rainfall Data (mm)")
print("==================")
print(f"Kampala: {kampala_rainfall}")
print(f"Gulu:    {gulu_rainfall}")

# Compute cosine similarity
similarity = cosine_similarity(kampala_rainfall, gulu_rainfall)

print(f"\nCosine Similarity: {similarity:.4f}")

# Interpretation
print("\nSimilarity Interpretation:")
if similarity >= 0.95:
    print("Very high similarity - Almost identical rainfall patterns")
elif similarity >= 0.85:
    print("High similarity - Very similar seasonal trends")
elif similarity >= 0.70:
    print("Moderate similarity - Somewhat similar patterns")
elif similarity >= 0.50:
    print("Low similarity - Different seasonal patterns")
else:
    print("Very low similarity - Very different rainfall behavior")
```

5.4 Use scipy.linalg.solve to model a simple supply-demand system.

```python
import numpy as np
from scipy import linalg

# Your data
passenger_counts = np.array([35, 40, 42, 50, 55, 60, 48, 52, 47, 45])
daily_revenue = np.array([70000, 80000, 84000, 100000, 110000, 120000, 96000, 104000, 94000, 90000])

# Calculate average price per passenger
average_price = np.mean(daily_revenue) / np.mean(passenger_counts)

# Simple supply-demand model:
# Demand: Qd = 100 - 0.8*P  (fewer passengers at higher prices)
# Supply: Qs = 20 + 0.6*P   (more service at higher prices)

# Convert to equations for scipy.linalg.solve:
# Q + 0.8P = 100  (from demand)
# Q - 0.6P = 20   (from supply)

A = np.array([[1, 0.8],   # Coefficients from demand equation
              [1, -0.6]]) # Coefficients from supply equation

b = np.array([100, 20])   # Constants from the equations

# Solve for equilibrium [Q, P]
equilibrium = linalg.solve(A, b)

print("SUPPLY-DEMAND EQUILIBRIUM")
print(f"Equilibrium passengers: {equilibrium[0]:.0f}")
print(f"Equilibrium price: UGX {equilibrium[1]:.0f}")
print(f"Equilibrium revenue: UGX {equilibrium[0] * equilibrium[1]:,.0f}")
print(f"The average actual price: UGX {average_price:.0f}")
```
